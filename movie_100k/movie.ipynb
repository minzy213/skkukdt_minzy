```python
avg_rating_df = rating_df.groupby('item_id').mean().sort_values('rating', ascending=False)
avg_rating_df
```

```python
top_rating_df = avg_rating_df[avg_rating_df.rating == 5].reset_index()
top_rating_df
```

```python
# movie_df
movie_df[movie_df['movie_id'].isin(top_rating_df['item_id'])]
```

```python
year_df = movie_df['release_date'].apply(lambda date : str(date).split('-')[-1])
```

```python
year_df
```

```python
# 이거를 쉽게!
movie_df['day'] = int(pd.DatetimeIndex(movie_df['release_date']).day)
movie_df['month'] = int(pd.DatetimeIndex(movie_df['release_date']).month)
movie_df['year'] = int(pd.DatetimeIndex(movie_df['release_date']).year)
```

```python
movie_df.sort_values(['year', 'month'], ascending=False)[['movie_title', 'year', 'month']]
```

### 미성년자(20살 미만)들이 좋아하는 영화
1. 미성년자 id 추출
1. 미성년자가 본 영화
1. 그 중에서 평균 평점이 5점인 영화(movie_id별로 평균)


```python
young_ids = user_df[user_df.age < 20].user_id
young_rating_df = rating_df[rating_df['user_id'].isin(young_ids)]
top_movie_df = young_rating_df.groupby('item_id').mean().sort_values('rating', ascending=False)
top_movie_ids = top_movie_df[top_movie_df['rating'] == 5].reset_index()
movie_df[movie_df['movie_id'].isin(top_movie_ids['item_id'])]
```
