```python
import pandas as pd
data = pd.read_csv('/content/drive/MyDrive/dataset/Ml_Challenge/Used_Cars_Price_Prediction.csv', index_col = 0)
data
```

```python
data.isna().sum()
```

```python
def getfirst(data):
    return data.split()[0].lower()

data['brand'] = data['Name'].apply(getfirst)
```

```python
data = data.drop('New_Price', axis = 1)
data = data.dropna()
data = data.drop('Name', axis = 1)
```

```python
# km/kg -> kmpl, tofloat
def mlgtoFloat(data):
    sp = data.split()
    ret = float(sp[0])
    if sp[1] == 'km/kg':
        ret *= 1.40
    return ret

data['Mileage'] = data['Mileage'].apply(mlgtoFloat)
data['Mileage']
```

```python
(data['Power'] == 'null bhp').sum()
```

```python
import numpy as np
def bhptoFloat(data):
    sp = data.split()
    if sp[0].isalpha():
        return np.NaN
    else:
        return float(sp[0])

data['Power'] = data['Power'].apply(bhptoFloat)
```

```python
data['Engine'] = data['Engine'].apply(lambda x : int(x.split()[0]))
```
