```python
# km/kg -> kmpl, tofloat
def mlgtoFloat(data):
    sp = data.split()
    ret = float(sp[0])
    if sp[1] == 'km/kg':
        ret *= 1.40
    return ret

data['Mileage'] = data['Mileage'].apply(mlgtoFloat)
data['Mileage']
```
